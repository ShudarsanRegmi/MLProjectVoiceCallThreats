```python
import nltk
import pickle
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Sample training data (expand this with real-world datasets)
training_data = [
    ("Hello, this is your bank. Please provide your OTP to verify your account.", 1),
    ("Congratulations! You have won a lottery. Provide details to claim your prize.", 1),
    ("Your social security number has been compromised. Act now!", 1),
    ("Dear customer, update your KYC to avoid account suspension.", 1),
    ("Hello, your Amazon order has been shipped and will arrive soon.", 0),
    ("Hi, just checking if you're available for dinner tonight.", 0),
    ("Reminder: Your doctor’s appointment is scheduled for tomorrow.", 0),
    ("Your package delivery is on the way.", 0),
]
```

```python
# Extract features and labels
texts, labels = zip(*training_data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
model = make_pipeline(TfidfVectorizer(), LogisticRegression())
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Save model to a file
pickle.dump(model, open("scam_model.pkl", "wb"))
print("✅ Scam detection model trained and saved!")
```

```python
# Evaluate model performance
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print performance metrics
print("\n📊 Model Performance:")
print(f"✔️ Accuracy: {accuracy:.2f}")
print(f"✔️ Precision: {precision:.2f}")
print(f"✔️ Recall: {recall:.2f}")
print(f"✔️ F1-score: {f1:.2f}")
```

```python
# Testing model on a sample input
test_input = ["You have won the lottery of 1 crore. Please click the link below to claim your reward."]
prediction = model.predict(test_input)[0]
print("\n📢 Test Prediction:")
print(f"Input: {test_input[0]}")
print(f"🔍 Predicted Scam: {'Yes' if prediction == 1 else 'No'}")
```
